## Probability using Two Fair Dice

Two fair and unbiased dice are rolled simultaneously.

- Each die has 6 faces
- Total outcomes = 6 Ã— 6 = 36
- Outcomes where both faces are same: (1,1), (2,2), ..., (6,6)

```python
# Number of faces on a die
faces = 6

# Total possible outcomes
total_outcomes = faces * faces

# Outcomes where both dice show the same face
same_face_outcomes = faces

# Probability of getting both faces same
probability = same_face_outcomes / total_outcomes

# Display results
print('Total number of outcomes:', total_outcomes)
print('Number of outcomes with both faces same:', same_face_outcomes)
print('Probability of getting both faces same:', probability)
```

### Conclusion
- Total outcomes = 36
- Favorable outcomes = 6
- Probability of getting both faces same = 1/6
